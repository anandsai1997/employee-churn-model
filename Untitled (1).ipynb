```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
```

```python
le_count=0
for col in df.columns[1:]:
    if df[col].dtype == 'object':
        if len(list(df[col].unique())) <= 2:
            le.fit(df[col])
            df[col] = le.transform(df[col])
            le_count += 1
print('{} columns were label encoded.'.format(le_count))

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 7))
df_col = list(df.columns)
df_col.remove('Attrition')
for col in df_col:
    df[col] = df[col].astype(float)
    df[[col]] = scaler.fit_transform(df[[col]])
df['Attrition'] = pd.to_numeric(df['Attrition'], downcast='float')
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df,
                                                    target,
                                                    test_size=0.25,
                                                    random_state=35,
                                                    stratify=target) 

print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
df.isnull().values.any()
```

```python
from sklearn.model_selection import GridSearchCV

rf_model = RandomForestClassifier(class_weight = "balanced",
                                       random_state=35)
rf_model.fit(X_train,y_train)

```

```python
rf_predict_train=rf_model.predict(X_train)
```

```python
from sklearn import metrics
```

```python
rf_predict_test=rf_model.predict(X_test)
print('Accuracy of RandomForest Regression Classifier on test set: {:.2f}'.format(metrics.accuracy_score(y_test,rf_predict_test)))
print()
```

```python
print("confusion matrix")
print("{0}".format(metrics.confusion_matrix(y_test,rf_predict_test)))
print("")
```

```python
print("classification report")
print(metrics.classification_report(y_test,rf_predict_test))
print("")
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier

forest = ExtraTreesClassifier(n_estimators=368,random_state=0)
forest.fit(X_train, y_train)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]


print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %d ((%f))"  %(f + 1, indices[f], importances[indices[f]]))   
 
plt.figure()
plt.title("Feature importances")
plt.bar(range(X_train.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X_train.shape[1]), indices)
plt.xlim([-1, X_train.shape[1]])
plt.show()

```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression(C=0.7,random_state=35)
lr_model.fit(X_train,y_train)
lr_predict_test=lr_model.predict(X_test)
```

```python
print("classification report")
print(metrics.classification_report(y_test,lr_predict_test))
print("")
```
